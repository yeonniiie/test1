```python
import pandas as pd
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
watch_data=[
    ['kbs2','tvn'],
    ['mbc','tvn'],
    ['jtbc','tvn'],
    ['kbs2','mbc'],
    ['kbs2','tvn','jtbc'],
    ['mbc','jtbc'],
    ['kbs2','tvn']
]
# tvn -> mbc
# 1/5
# 3/7
```

```python
(1/5) / (3/7)
```

```python
# 인코딩
te=TransactionEncoder()
te_ary=te.fit(watch_data).transform(watch_data)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
freq =  apriori(df, min_support=0.1, use_colnames=True)
# 예초기 -> 낫
# 전체 구매에서 낫 대비 예초기 중 낫
# (4/5) / (10/1000)
```

```python
freq
```

```python
rules = association_rules(freq, metric='confidence',min_threshold=0.3)
```

```python
rules
```

```python
rules['source'] = rules['antecedents'].apply(lambda x: ','.join(list(x)))
rules['target'] = rules['consequents'].apply(lambda x: ','.join(list(x)))
```

```python
rules
```

```python
sdf = rules[['source','target','lift']]
sdf.columns = ['source','target','lift']
```

```python
sdf
```

```python
sdf.to_csv('lg_hellovision.csv',index=False)
```
